````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# URL'yi temizleyelim
url = "https://newyork.craigslist.org/search/cta?query=toyota"

# TarayÄ±cÄ± gibi gÃ¶rÃ¼nmek iÃ§in daha detaylÄ± bir header ekleyelim
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36",
    "Accept-Language": "tr-TR,tr;q=0.9,en-US;q=0.8,en;q=0.7"
}

response = requests.get(url, headers=headers)

# EÄŸer sayfa yÃ¼klenmediyse hata mesajÄ± basalÄ±m
if response.status_code != 200:
    print(f"Hata: Sayfaya eriÅŸilemedi. Durum kodu: {response.status_code}")
else:
    soup = BeautifulSoup(response.text, "html.parser")

    # Craigslist yeni yapÄ±sÄ±nda ilanlar genelde 'li' iÃ§inde 'cl-static-search-result' class'Ä±na sahip
    # Veya doÄŸrudan baÅŸlÄ±k ve fiyat etiketlerini arayabiliriz
    cars_elements = soup.find_all("li", class_="cl-static-search-result")

    print(f"Ä°lan sayÄ±sÄ±: {len(cars_elements)}")

    cars = []

    for car in cars_elements:
        # BaÅŸlÄ±k ve Link
        title_tag = car.find("div", class_="title")
        title = title_tag.text if title_tag else "N/A"

        # Fiyat
        price_tag = car.find("div", class_="price")
        price = price_tag.text if price_tag else "N/A"

        # Lokasyon
        location_tag = car.find("div", class_="location")
        location = location_tag.text.strip() if location_tag else "N/A"

        cars.append([title, price, location])

    # DataFrame oluÅŸturma
    df = pd.DataFrame(cars, columns=["Title", "Price", "Location"])
    print(df.head())

    # Veriyi kaydetmek isterseniz:
    # df.to_csv("toyota_ilanlar.csv", index=False)
    display(df)
````

### Daha geniÅŸ kapsamlÄ± bir veri Ã§ekme iÅŸlemi yapalÄ±m.

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time

# --------------------------------------------------
# AYARLAR
# --------------------------------------------------
url = "https://newyork.craigslist.org/search/cta?query=toyota"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
}

# --------------------------------------------------
# Ä°LAN LÄ°NKLERÄ°NÄ° TOPLA
# --------------------------------------------------
print("Linkler toplanÄ±yor...")
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

links = [
    a["href"]
    for a in soup.select("li.cl-static-search-result a")
    if a.get("href")
]

print(f"{len(links)} adet ilan bulundu.")

# --------------------------------------------------
# Ä°LAN DETAYLARINI Ã‡EK
# --------------------------------------------------
all_cars = []

for i, link in enumerate(links[:100]):  # ilk 100 ilan
    try:
        print(f"{i + 1}. ilan detaylarÄ± okunuyor...")
        res = requests.get(link, headers=headers)
        s = BeautifulSoup(res.text, "html.parser")

        car_info = {"Link": link}

        # BaÅŸlÄ±k
        title_tag = s.find("span", id="titletextonly")
        car_info["Title"] = title_tag.text if title_tag else np.nan

        # Fiyat
        price_tag = s.find("span", class_="price")
        car_info["Price"] = (
            price_tag.text.replace("$", "").replace(",", "")
            if price_tag else np.nan
        )

        # Ã–zellikler (attrgroup)
        for attr in s.select(".attrgroup .attr"):
            label = attr.find(class_="label")
            value = attr.find(class_="value")

            if label and value:
                key = label.text.replace(":", "").strip().lower()
                val = value.text.strip()
                car_info[key] = val
            elif ":" in attr.text:
                parts = attr.text.split(":", 1)
                car_info[parts[0].strip().lower()] = parts[1].strip()

        all_cars.append(car_info)
        time.sleep(1)

    except Exception as e:
        print("Hata:", e)
        continue

# --------------------------------------------------
# DATAFRAME OLUÅTUR
# --------------------------------------------------
df = pd.DataFrame(all_cars)

# --------------------------------------------------
# TEMÄ°ZLÄ°K
# --------------------------------------------------
# BoÅŸluk / None â†’ NaN
df = df.replace(r'^\s*$', np.nan, regex=True)

# ğŸ”¹ TITLE'DAN ARAÃ‡ YILI Ã‡IKAR
df["year"] = df["Title"].str.extract(r'(19\d{2}|20\d{2})')
df["year"] = pd.to_numeric(df["year"], errors="coerce")

# --------------------------------------------------
# SADECE GEREKLÄ° SÃœTUNLAR
# --------------------------------------------------
istenenler = [
    "Title",
    "year",
    "Price",
    "condition",
    "cylinders",
    "drive",
    "fuel",
    "odometer",
    "paint color",
    "transmission",
    "type"
]

df_final = df.reindex(columns=istenenler)

# --------------------------------------------------
# SAYISAL DÃ–NÃœÅÃœMLER
# --------------------------------------------------
df_final["Price"] = pd.to_numeric(df_final["Price"], errors="coerce")

df_final["odometer"] = (
    df_final["odometer"]
    .str.replace(",", "", regex=False)
    .astype(float)
)

# MantÄ±ksÄ±z yÄ±llarÄ± temizle
df_final.loc[
    (df_final["year"] < 1980) | (df_final["year"] > 2025),
    "year"
] = np.nan

# --------------------------------------------------
# SONUÃ‡
# --------------------------------------------------
print("\nTablo GÃ¼ncellendi!")
display(df_final.head(10))

# --------------------------------------------------
# CSV KAYDET
# --------------------------------------------------
df_final.to_csv(
    "toyota_crap_data.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Veri seti 'toyota_crap_data.csv' olarak kaydedildi!")
````
